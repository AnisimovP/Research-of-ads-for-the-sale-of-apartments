# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

# Описание данных

* `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
* `balcony` — число балконов
* `ceiling_height` — высота потолков (м)
* `cityCenters_nearest` — расстояние до центра города (м)
* `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
* `first_day_exposition` — дата публикации
* `floor` — этаж
* `floors_total` — всего этажей в доме
* `is_apartment` — апартаменты (булев тип)
* `kitchen_area` — площадь кухни в квадратных метрах (м²)
* `last_price` — цена на момент снятия с публикации
* `living_area` — жилая площадь в квадратных метрах (м²)
* `locality_name` — название населённого пункта
* `open_plan` — свободная планировка (булев тип)
* `parks_around3000` — число парков в радиусе 3 км
* `parks_nearest` — расстояние до ближайшего парка (м)
* `ponds_around3000` — число водоёмов в радиусе 3 км
* `ponds_nearest` — расстояние до ближайшего водоёма (м)
* `rooms` — число комнат
* `studio` — квартира-студия (булев тип)
* `total_area` — площадь квартиры в квадратных метрах (м²)
* `total_images` — число фотографий квартиры в объявлении

### Обработка пропусков

#### Балконы

```python
data['balcony'].value_counts()
```

```python
data['balcony'].isna().sum()
```

```python
data['balcony'] = data['balcony'].fillna(0)
```

```python
data['balcony'].value_counts()
```

В столбце `balcony` заменили пропуски на значение "0" т.к. скорее всего, если оставили пропуск, имеется ввиду, что в квартире нету балконов.

#### Комнтаы

```python
data['rooms'].describe()
```

В столбце `rooms` были замечаны значения "0", что естественно не может быть. (в квартире не может быть 0 комнат) Возможно, в квартире одна большая студия, и клиент, который заполнял данные не посчитал это за комнату в стандартном понимании комнаты. Так что заменим значения "0" на "1"

```python
data['rooms'] = data['rooms'].replace(0, 1)
```

```python
data['rooms'].value_counts()
```

#### Высота потолков

```python
data['ceiling_height'].describe()
```

```python
data['ceiling_height'].isna().sum()
```

```python
median_ceiling_height = data['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(median_ceiling_height)
```

```python
data['ceiling_height'].isna().sum()
```

В стобце `ceiling_height` были найдены пропуски данных (наверное пользователь просто забыл указать высоту потолков). Так что заменили пропуски медианным значением по всему столбцу, чтобы в дальнейшем иметь возможность производить подсчеты.

#### Количество этажей

```python
data['floors_total'].describe()
```

```python
data['floors_total'].isna().sum()
```

```python
data.isna().mean()
```

```python
data = data.dropna(subset=['floors_total'])
```

```python
data['floors_total'].isna().sum()
```

В стобце `floors_total` были найдены пропуски данных (наверное пользователь просто забыл указать количество этажей). Так как эти данные участвуют в анализе и они сильно влияют на один пункт(нельзя точно сказать человек живет на последнем этаже в доме или нет, заменив к примеру медианным значением, средним, каким-то максимальным по данному столбцу), а пропущенные значения занимают меньше половина процента от всех значений, было принято решение, удалить строки с пропущенными значениями.

```python
data['first_day_exposition']
```

```python
df = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
data['weekday_publication'] = df.dt.weekday
data['month_publication'] = df.dt.month
data['year_publication'] = df.dt.year
data.head()
```

```python
def weekday(week):
    if week == 0:
        return 'Понедельник'
    if week == 1:
        return 'Вторник'
    if week == 2:
        return 'Среда'
    if week == 3:
        return 'Четверг'
    if week == 4:
        return 'Пятница'
    if week == 5:
        return 'Суббота'
    if week == 6:
        return 'Воскресенье'
```

```python
data['weekday_publication'] = data['weekday_publication'].apply(weekday)
```

```python
def month_count(month):
    if month == 1:
        return 'Январь'
    if month == 2:
        return 'Февраль'
    if month == 3:
        return 'Март'
    if month == 4:
        return 'Апрель'
    if month == 5:
        return 'Май'
    if month == 6:
        return 'Июнь'
    if month == 7:
        return 'Июль'
    if month == 8:
        return 'Август'
    if month == 9:
        return 'Сентябрь'
    if month == 10:
        return 'Октбярь'
    if month == 11:
        return 'Ноябрь'
    if month == 12:
        return 'Декабрь'
```

```python
data['month_publication'] = data['month_publication'].apply(month_count)
```

```python
data['price_per_meter'] = round(data['last_price'] / data['total_area'], 2)
data.head(10)
```

### Этаж квартиры

```python
def floor_flat(floor):
    
    floors_total = floor['floors_total']
    floor_living = floor['floor']
    if floor_living == 1:
        return 'Первый'
    if floor_living == floors_total:
        return 'Последний'
    else:
        return 'Другой'
```

```python
data['floor_flat'] = data.apply(floor_flat, axis=1)
```
